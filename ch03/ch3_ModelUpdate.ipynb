```python
import numpy as np
import matplotlib.pyplot as plt

hours = np.array([10, 7, 5, 4])
result = np.array([1, 1, 0, 0])
```

```python
def LinR(x, w=2, b=-10):
    return w * x + b

print(LinR(hours))
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def LogR(x, w=2, b=-10):
    z = LinR(x, w, b)
    h_z = sigmoid(z)
    return h_z

print(LogR(hours))
```

```python
x = np.linspace(-1, 12, 100)
pred = LogR(x)
plt.plot(x, pred)
plt.scatter(hours, result)
```

```python
def costfunction(h, y):
    return -(y*np.log(h)+(1-y)*np.log(1-h))

pred = LogR(hours)

print(pred)
print(np.log(pred))                   #np.log = 밑이 e인 자연로그 = ln
print(costfunction(pred, np.array(result)))
```

```python
lr = 2.0
w = 2
b = -10

new_b = b - lr*np.mean(pred - result)
print(new_b)

new_w = w - lr*np.mean(pred - result)*(hours)
print(new_w)
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def new_LogR(x, w=2, b=-10.3):
    z = LinR(x, w, b)
    h_z = sigmoid(z)
    return h_z

print(new_LogR(hours))
```

```python
x = np.linspace(-1, 12, 100)

b = -10.0
pred = LogR(x)
plt.plot(x, pred)

new_b = -10.3
new_pred = new_LogR(x)
plt.plot(x, new_pred)

plt.scatter(hours, result)
plt.legend()
```
