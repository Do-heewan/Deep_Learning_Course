```python
import numpy as np
import matplotlib.pyplot as plt

L = np.array([4, 8, 12, 16, 20])
Time = np.array([2, 4, 6, 8, 10])
```

```python
plt.scatter (L, Time)
plt.xlim(0,30)
plt.xlabel('L')

plt.ylim(0,15)
plt.ylabel('Time')
```

```python
def LR(x, w = 0.25, b = 0):
    return w*x + b

L_line = np.linspace(0, 28, 5)
h = LR(L_line)
print(h)
```

```python
plt.plot(L_line, h, 'r')
plt.scatter(L, Time)

plt.xlim(0,30)
plt.xlabel('L')

plt.ylim(0,15)
plt.ylabel('Time')
```

```python
def cost(y, y_hat):
    return np.array((y - y_hat) ** 2)

h = LR(L_line)
loss = cost(Time, h)
print(loss)
```

```python
w = 0.25
b = 0

lr = 0.002

n = 5

h = LR(L, w=w ,b=b)
dW = (2/n) * sum(L * (h - Time))
w_new = w - lr * dW

print(dW)
print(w_new)
```
