```python
import numpy as np
import matplotlib.pyplot as plt

score = np.array([25, 49, 42, 57, 72, 68])
hours = np.array([2, 3, 4, 5, 6, 7])
```

```python
def LR(x, w=3, b=75):
  return wx+b
```

```python
plt.scatter (hours, score)
plt.xlim(0,10)
plt.xlabel('hours')

plt.ylim(0,100)
plt.ylabel('score')

plt.title('hours Vs score')
```

```python
def model(x):
  w=3
  b=35
  return w*x+b

hours_0to10 = np.linspace(0, 10, 11)
h = model(hours_0to10)
print(h)
```

```python
plt.plot(hours_0to10, h, 'r')
plt.scatter(hours, score)
plt.xlim(0,10)
plt.ylim(0,100)
```

```python
def model (x, w=3, b=35):
  return w*x + b

for i in range(-10,20,1):
  h = model(hours_0to10, w=i)
  plt.plot(hours_0to10, h)

plt.scatter(hours, score)
plt.xlim(0,10)
plt.ylim(0,100)
```

```python
def cost (y, y_hat):
    return np.mean((y - y_hat) ** 2)

h = model(hours)
loss = cost(score, h)
print(loss)
```

```python
for i in range (-20, 30, 1):
    h = model(hours, w=i)
    loss = cost(score, h)
    plt.scatter(i, loss)
```

```python
for i in range(-10,90,4):
  h = model(hours_0to10, b=i)
  plt.plot(hours_0to10,h)

plt.scatter(hours, score)
plt.xlim(0,10)
plt.ylim(0,100)
```

```python
for i in range (-10, 90, 4):
    h = model(hours, b = i)
    loss = cost(score, h)
    plt.scatter(i, loss)
```

```python
w = np.random.uniform()
b = np.random.uniform()

w_init = w
b_init = b

print(w)
print(b)
```

```python
#hyperparameter
lr = 0.001
epochs = 10000
```

```python
n = hours.shape[0]

loss = []

for i in range(epochs):
  h = model(hours, w=w, b=b)

  loss.append(cost(score, h))

  dw = (2/n) * sum(hours * (h - score)) # np.mean(hours * (h - score))
  db = (2/n) * sum(h - score)           # np.mean(h - score)

  w = w - lr * dw
  b = b - lr * db

print(loss[-1])
```

```python
plt.plot(loss)
plt.ylim(0,300)
```
